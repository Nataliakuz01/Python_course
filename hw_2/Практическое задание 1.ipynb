```python
class Chocolate:
  def __init__ (self, price, add=None):
    self.price = float (price)
    self.add = str (add)
    if price <= 0:
      raise ValueError ('Ошибка! Цена меньше 0')


  def show_my_chocolate (self):
      if self.add == 'None':
        print ('Обычный шоколадный батончик')
      else:
        print (f'Шоколадный батончик, добавка: {self.add}')

  def __str__(self):
      return f'Шоколадный батончик, {self.price} рублей'
```

```python
Mars = Chocolate(169.99, 'Фундук')
Mars.show_my_chocolate()
print(Mars)
```

```python
Snickers = Chocolate(100)
print (Snickers)
Snickers.show_my_chocolate()
```
